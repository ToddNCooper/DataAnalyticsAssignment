From the analysis of the job-market dataset, the findings show that - if an individual is looking for the most job opportunity, Sydney, followed by Melbourne, then Brisbane 
are the top cities in Australia with the most job listings. Sydney accounting for 27.57% of the job market, Melbourne accounting for 24.56%, and Brisbane accounting for 11.08%. Regardless of the job classification chosen, these cities still provide the highest variety and number listings for each job type. 

When looking at the median salary for each location the results show a different story. Sydney, Melbourne, and Brisbane are not even in the top 10. The location with the highest median salary of 43.0 is Whyalla & Eyre Peninsula. When including locations with less than 10 job listings this ranking changes, but these low count locations were considered as outliers due to the unfair nature when analysing mean values. Whyalla & Eyre Peninsula would be the best job location when considering highest median salary. Locations to avoid would include Kalgorile, Goldfeilds & Esperance, Port Hedland, Karratha & Pilbara. These locations have a mediam salary of around 24.0-27.0.

The hottest job sectors in the top 3 cities are Hospitality & Tourism, Trades & Services, and Manufacturing, Transport & Logistics. These sectors populate a large majority of the market. An inidividual currently in the market for a job would have the best change at securing a job in these sectors due to the large availability across found in the largest cities.
